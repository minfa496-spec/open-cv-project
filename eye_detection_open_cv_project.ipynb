```python
# Step 1: Import libraries
import cv2
from google.colab.patches import cv2_imshow

# Step 2: Load the Haar cascade for eyes only
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Step 3: Read the image file (replace 'your_image.jpg' with your file)
img = cv2.imread('/content/WhatsApp Image 2025-09-24 at 3.32.03 PM.jpeg')

# Step 4: Convert image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 5: Detect eyes directly
eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

for (ex, ey, ew, eh) in eyes:
    cv2.rectangle(img, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)

# Step 6: Show the result
cv2_imshow(img)
cv2.waitKey(0)  # Wait until a key is pressed
cv2.destroyAllWindows()
```
